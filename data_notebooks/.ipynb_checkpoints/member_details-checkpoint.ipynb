```python
# --- Cell H1: Load hear/volunteer CSV and detect columns ---

import pandas as pd
from pathlib import Path

HEAR_CSV = Path("../data/membership/member_hear.csv")

hear = pd.read_csv(HEAR_CSV, dtype=str)
hear.columns = hear.columns.str.strip()

print("Rows:", len(hear))
print("Columns:", hear.columns.tolist())

# Helpers to find columns by substrings (case-insensitive)
def find_col(df, candidates, required=True):
    for c in df.columns:
        low = c.lower()
        if any(t in low for t in candidates):
            return c
    if required:
        raise KeyError(f"Could not find any of: {candidates}")
    return None

# Detect likely columns
HEAR_ID   = find_col(hear, ["account id", "account_id", "id"], required=False)
HEAR_SRC  = find_col(hear, ["how did you hear", "how heard", "hear about", "referral source", "source"], required=False)
HEAR_VOL  = find_col(hear, ["volunteer", "committee", "help with"], required=False)
HEAR_ZIP  = find_col(hear, ["zip"], required=False)
HEAR_EMAIL= find_col(hear, ["email"], required=False)

print("\nDetected:")
print("  Account ID  :", HEAR_ID)
print("  How-Heard   :", HEAR_SRC)
print("  Volunteer   :", HEAR_VOL)
print("  ZIP         :", HEAR_ZIP)
print("  Email       :", HEAR_EMAIL)

# Show quick value counts (top 15) for detected cols
for label, col in [("How-Heard", HEAR_SRC), ("Volunteer", HEAR_VOL)]:
    if col and col in hear.columns:
        print(f"\nTop values for {label} [{col}]:")
        vc = hear[col].fillna("(blank)").astype(str).str.strip()
        display(vc.value_counts().head(15).to_frame("count"))
    else:
        print(f"\n{label}: not found")
        
# Peek a few rows
display(hear.head(5))

df.info()
```

```python
# --- Cell H2: Clean + summarize How-Heard and Volunteer fields ---

import numpy as np
import matplotlib.pyplot as plt

# Exact columns from your df.info()
COL_ID      = "Account ID"
COL_LEVEL   = "Membership Level"
COL_HEARD   = "How did you hear about The Virginia Council? (C)"
COL_VOL     = "Would you like to volunteer for one of our committ (C)"

# 1) Basic completeness
n = len(df)
heard_nonnull = df[COL_HEARD].notna().sum()
vol_nonnull   = df[COL_VOL].notna().sum()
print(f"Rows: {n}")
print(f"How-Heard non-null: {heard_nonnull} ({heard_nonnull/n:.1%})")
print(f"Volunteer non-null: {vol_nonnull} ({vol_nonnull/n:.1%})")

# 2) Bucket How-Heard into consistent categories
def bucket_reason(x):
    if not isinstance(x, str) or not x.strip():
        return "unknown"
    s = x.strip().lower()
    s = s.replace("&", " and ").replace("/", " ").replace("-", " ")
    # Keyword buckets (mutually exclusive, ordered)
    if any(k in s for k in ["event", "meeting", "webinar", "conference"]):
        return "events"
    if any(k in s for k in ["referral", "friend", "colleague", "word of mouth"]):
        return "referral"
    if any(k in s for k in ["twitter", " x ", "linkedin", "facebook", "instagram", "social"]):
        return "social"
    if any(k in s for k in ["press", "media", "newspaper", "article", "news"]):
        return "press/media"
    if any(k in s for k in ["email", "newsletter", "mailing"]):
        return "email"
    if any(k in s for k in ["google", "search", "website", "web site", "site", ".org", ".com"]):
        return "web/search"
    return "other"

df["_ReasonBucket"] = df[COL_HEARD].astype(str).apply(bucket_reason)

# 3) Tidy summaries
reason_counts = (df["_ReasonBucket"]
                 .value_counts(dropna=False)
                 .rename("Count")
                 .to_frame())
reason_share  = (reason_counts["Count"] / n).rename("Share").to_frame()
reason_table  = reason_counts.join(reason_share).sort_values(["Count","Share"], ascending=[False, False])

# Volunteer: top answers (treat blanks as "(blank)")
vol_counts = (df[COL_VOL].fillna("(blank)")
              .astype(str).str.strip()
              .replace({"": "(blank)"})
              .value_counts()
              .rename("Count")
              .to_frame())
vol_share  = (vol_counts["Count"] / n).rename("Share").to_frame()
vol_table  = vol_counts.join(vol_share).sort_values(["Count","Share"], ascending=[False, False])

# Cross: Reason × Level
cross_reason_level = (df.pivot_table(index="_ReasonBucket",
                                     columns=COL_LEVEL,
                                     values=COL_ID,
                                     aggfunc="nunique",
                                     fill_value=0)
                      .sort_values(by=list(df[COL_LEVEL].unique()), ascending=False, axis=0))

# 4) Display tables
print("\n== How-Heard (bucketed) ==")
display(reason_table)

print("\n== Volunteer answers (top) ==")
display(vol_table.head(20))

print("\n== How-Heard × Membership Level (counts) ==")
display(cross_reason_level)

# 5) Quick plots
# How-Heard bar
ax = reason_table["Count"].plot(kind="bar")
ax.set_title("How-Heard (Bucketed) — Counts")
ax.set_xlabel("Reason")
ax.set_ylabel("Members")
plt.xticks(rotation=30, ha="right")
plt.tight_layout()
plt.show()

# Volunteer bar (top 12)
ax = vol_table["Count"].head(12).plot(kind="bar")
ax.set_title("Volunteer Interest — Top Responses")
ax.set_xlabel("Volunteer Response")
ax.set_ylabel("Members")
plt.xticks(rotation=30, ha="right")
plt.tight_layout()
plt.show()
```

```python
# --- Exact How-Heard responses: counts + renewal performance (12m) ---

import numpy as np
import matplotlib.pyplot as plt

# Columns in member_hear.csv dataframe `df`
COL_ID    = "Account ID"
COL_HEARD = "How did you hear about The Virginia Council? (C)"
COL_LEVEL = "Membership Level"  # present in member_hear.csv (optional for cross-tabs)

# 1) Normalize the exact answer text (no bucketing)
hear_exact = df[[COL_ID, COL_HEARD]].copy()
hear_exact[COL_HEARD] = hear_exact[COL_HEARD].fillna("(blank)").astype(str).str.strip()

# Preserve a useful display order (descending by count)
counts = hear_exact[COL_HEARD].value_counts().rename("Count")
order  = counts.index.tolist()

# 2) Plot: raw counts of each exact response
display(counts.to_frame())

ax = counts.plot(kind="bar")
ax.set_title("How did you hear about The Virginia Council? — Exact Responses")
ax.set_xlabel("Response")
ax.set_ylabel("Count")
plt.xticks(rotation=30, ha="right")
plt.tight_layout()
plt.show()

# 3) Renewal within ~12 months by exact response
# Recompute next enrollment on the history table `d` (from member_report.csv) to be safe
# (Assumes d has H_ID, H_ENROLL from earlier cells.)
# If not in memory, run the earlier 'history load + first_joins' cells first.
from pandas.api.types import is_datetime64_any_dtype as is_dt

# Detect column names used in earlier cells (history)
# If you used the exact earlier code, these exist:
H_ID_candidates = [c for c in d.columns if c.lower() in ("account id", "account_id")]
H_ENROLL_candidates = [c for c in d.columns if "enrollment" in c.lower()]
if not H_ID_candidates or not H_ENROLL_candidates:
    raise KeyError("History table `d` is missing expected columns. Re-run earlier history cells.")
H_ID, H_ENROLL = H_ID_candidates[0], H_ENROLL_candidates[0]

if not is_dt(d[H_ENROLL]):
    d[H_ENROLL] = pd.to_datetime(d[H_ENROLL], errors="coerce")

# next enrollment per account
d = d.sort_values([H_ID, H_ENROLL]).copy()
d["_next_enroll"] = d.groupby(H_ID)[H_ENROLL].shift(-1)

# Build or reuse first_joins consistently
if " _row_num_for_acct" in first_joins.columns or "_row_num_for_acct" in d.columns:
    pass
else:
    d["_row_num_for_acct"] = d.groupby(H_ID).cumcount() + 1
first_joins = d[d["_row_num_for_acct"].eq(1)].copy()

# 12-month eligibility anchor = latest enroll date in dataset
anchor = d[H_ENROLL].max()
if pd.isna(anchor):
    anchor = pd.Timestamp.today().normalize()

first_joins["_eligible12m"] = first_joins[H_ENROLL] <= (anchor - pd.Timedelta(days=365))
first_joins = first_joins.merge(
    d[[H_ID, H_ENROLL, "_next_enroll"]],
    on=[H_ID, H_ENROLL],
    how="left"
)
first_joins["_renewed12m"] = (
    first_joins["_eligible12m"]
    & first_joins["_next_enroll"].notna()
    & ((first_joins["_next_enroll"] - first_joins[H_ENROLL]).dt.days.between(0, 395, inclusive="both"))
)

# 4) Merge exact how-heard onto the first-join record
fj_hear = first_joins.merge(
    hear_exact[[COL_ID, COL_HEARD]],
    left_on=H_ID, right_on=COL_ID, how="left"
)
fj_hear[COL_HEARD] = fj_hear[COL_HEARD].fillna("(blank)").astype(str).str.strip()

# 5) Aggregate: by exact response
tbl = (
    fj_hear.groupby(COL_HEARD)
           .agg(
               First_Joins  = (H_ID, "nunique"),
               Eligible_12m = ("_eligible12m", "sum"),
               Renewed_12m  = ("_renewed12m", "sum"),
           )
           .reindex(order)  # keep display order by raw counts
)
tbl["Eligible_12m"] = tbl["Eligible_12m"].astype(int)
tbl["Renewed_12m"]  = tbl["Renewed_12m"].astype(int)
tbl["Renewal_Rate_12m"] = np.where(
    tbl["Eligible_12m"] > 0,
    tbl["Renewed_12m"] / tbl["Eligible_12m"],
    np.nan
)

print("== Renewal within ~12 months by exact How-Heard response ==")
display(tbl)

# 6) Plot: renewal rate by exact response (annotate Eligible N)
plot_data = tbl.dropna(subset=["Renewal_Rate_12m"]).copy()
if not plot_data.empty:
    ax = plot_data["Renewal_Rate_12m"].plot(kind="bar")
    ax.set_title("12-Month Renewal Rate by Exact How-Heard Response")
    ax.set_xlabel("Response")
    ax.set_ylabel("Renewal Rate")
    plt.xticks(rotation=30, ha="right")
    # annotate eligible counts
    for p, n in zip(ax.patches, plot_data["Eligible_12m"].values):
        ax.text(p.get_x() + p.get_width()/2, p.get_height() + 0.01, f"N={n}",
                ha="center", va="bottom", fontsize=9)
    plt.ylim(0, min(1.0, max(0.1, plot_data["Renewal_Rate_12m"].max() + 0.1)))
    plt.tight_layout()
    plt.show()
else:
    print("No categories with eligible cohorts for 12-month renewal yet (data may be too recent).")
```
